# Pandas:-

## Series:- 1-D

```python
import pandas as pd
```

```python
a=(10,20,30)
b=pd.Series(a)
b
```

```python
a=(10,20,30)
index=("riya","riyu","sarva")
b=pd.Series(index,index=a)
b
```

```python
name=[100,200,300]
ans=pd.Series(name)
print(ans)
```

```python
data={"a":10,"b":20,"c":30}
answer=pd.Series(data)
answer
```

```python
data={"a":10,"b":20,"c":30}
answer=pd.Series(data,name="score")
answer
```

## DataFrame:-

```python
data=[[1,2,3],[4,5,6],[7,8,9]]
df=pd.DataFrame(data)
print(df)
```

```python
data=[[1,2,3],[4,5,6],[7,8,9]]
df=pd.DataFrame(data)
df
```

```python
data=[[1,2,3],[4,5,6],[7,8,9]]
df=pd.DataFrame(data,columns=["D1","D2","D3"])
df
```

```python
data=[[1,2,3],[4,5,6],[7,8,9]]
df=pd.DataFrame(data,columns=["D1","D2","D3"],index=["r1","r2","r3"])
df
```
