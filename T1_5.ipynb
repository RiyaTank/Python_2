```python
import networkx as nx
import matplotlib.pyplot as plt


# Initial dictionary provided
instagram = {
    'person1': [0, 1, 1, 0, 1],
    'person2': [0, 0, 1, 0, 1],
    'person3': [1, 1, 0, 1, 1],
    'person4': [1, 1, 1, 0, 0],
    'person5': [1, 1, 0, 0, 0]
}

# 1. Initialize a Directed Graph object
G = nx.DiGraph()

# 2. Map the keys to a list to use indices for edge creation
people = list(instagram.keys())

# 3. Iterate through the dictionary to add directed edges
for follower, follows_list in instagram.items():
    for index, is_following in enumerate(follows_list):
        if is_following == 1:
            # Add edge from the 'follower' to the 'followee' at the current index
            G.add_edge(follower, people[index])

# Check the results
print(f"Nodes: {G.nodes()}")
print(f"Edges: {G.edges()}")
plt.show()
```
